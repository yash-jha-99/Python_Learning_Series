# MatplotLib tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
# Simple examples
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
##Ploting using matplotLib
##plt scatter

plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Graph in 2D")
plt.savefig('image1.jpeg')
```

```python
## plt plot
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=1,markersize=10)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2d Diagram')
```

```python
## Creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,4)
plt.plot(y,x,'y--')
plt.subplot(2,2,3)
plt.plot(y,x,'go')
```

```python
# more examples
x = np.arange(1,11) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel=("x axis caption") 
plt.ylabel=("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
# compute the x and y coordinate for points on a site curve
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')

#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#subplot
# compute the x and y coordinates for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine')
```

```python
## Bar plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

Box plot using matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
plt.title("Boxplot")
plt.show()
      
```

Pie chart

```python
labels='Python','java','Ruby','C++'
sizes=[225,130,245,210]
colors=['gold','Yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #exploode 1st slice

#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.2f%%',shadow=True)
plt.axis('equal')
plt.show()
```
