```python
type(data)
```

```python
StringIO()
```

```python
##Index columns and training delimaters
```

```python
data=('index,a,b,c\n'
         '4,apple,bat,5\n'
         '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
data=('a,b,c\n'
         '4,apple,bat,\n'
         '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
## Combining usecols and index_col
data=('a,b,c\n'
         '4,apple,bat,\n'
         '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data),usecols=['b','c'],index_col=False)
```

```python
## Quoting and escape character. Very useful in NLP
data='a,b\n "Hello  \\"bob\\"\\, nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```

```python
## URL to read csv
## If we have to read csv data from urls and it seperated byu \t the we have to use function sep='\t' after url
pd.read_csv("https://support.staffbase.com/hc/en-us/article_attachments/360009197011/username-password-recovery-code.csv",sep='\t')
```
