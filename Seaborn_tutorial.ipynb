### Correlation with heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D Correlation matrix (table) between two discrete dimensions or event types, It is very important in Feature selection.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## JoinPlot

A joinPlot allows the study of relationship between 2 numeric variables. The central chart display the correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

### Univariate Analysis 

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pair Plot

A "pairs plot" is also known as Scatterplot. In which one variable in the same data row is matched with another variable's value, like this: pairs plots are just elaborations on this., showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

### Dist Plot 

Dist plot helps us to check with the distribution of the column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde='False',bins=10)
```

# Categorical Plots

 ##### Seaborn also helps us to doing the analysis of categorical Data points.
 
 
   -  boxplot
   -  violinplot
   -  countplot
   -  bar plot

```python
## count plot

sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
## Bar plot

sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
## Box plot
## A box and whisker plot(sometimes called a box plot) is a graph that presents information from a five-number summary 
```

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
# Violin plot
# Violin plot helps us to see both the distribution of data in terms of kermnal density and box plot
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
